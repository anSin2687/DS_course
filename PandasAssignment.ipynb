```python
#ANS 16 OPTION D ans is data.apply(functionName,axis=1)
def GradeChange(data):
    if data.sqft_living>0 and data.sqft_living<=400:
        return data.grade+1
    elif data.sqft_living>400:
        return data.grade+2
    
        
data["New_Grade"] = data.apply(GradeChange,axis=1)
data[['New_Grade','grade']]
```
